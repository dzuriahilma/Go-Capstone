```python
# Daftar kolom yang ingin diisi dengan forward fill
columns_to_ffill = ['Brand', 'Price', 'Rating', 'Ingredients', 'Combination', 'Dry', 'Normal', 'Oily', 'Sensitive', 'Category', 'Product_Url', 'User_id', 'Product_id', 'Ingredients_Cleaned', 'Review_Cleaned', 'Good_Stuff', 'Ing_Tfidf']

# Mengisi missing values dengan forward fill untuk kolom yang ditentukan
data[columns_to_ffill] = data[columns_to_ffill].fillna(method='ffill')

print(data)

```

```python
for i in range(len(data['Ingredients_Cleaned'])):
    data['Ingredients_Cleaned'].iloc[i] = str(data['Ingredients_Cleaned'].iloc[i]).replace('[', '').replace(']', '').replace("'", '').replace('"', '')
```

```python
all_ingreds = []

for i in data['Ingredients_Cleaned']:
    ingreds_list = i.split(', ')
    for j in ingreds_list:
        all_ingreds.append(j)
```

```python
all_ingreds = sorted(set(all_ingreds))
all_ingreds[0:20]
```

```python
one_hot_list = [[0] * 0 for i in range(len(all_ingreds))]

for i in data['Ingredients_Cleaned']:
    k=0
    for j in all_ingreds:
        if j in i:
            one_hot_list[k].append(1)
        else:
            one_hot_list[k].append(0)
        k+=1

ingred_matrix = pd.DataFrame(one_hot_list).transpose()
ingred_matrix.columns = [sorted(set(all_ingreds))]

ingred_matrix
```

```python
svd = TruncatedSVD(n_components=150, n_iter = 1000, random_state = 6) # firstly reduce features to 150 with truncatedSVD - this suppresses some noise
svd_features = svd.fit_transform(ingred_matrix)
tsne = TSNE(n_components = 2, n_iter = 1000000, random_state = 6) # reduce 150 features to 2 using t-SNE with exact method
tsne_features = tsne.fit_transform(svd_features)

data['X'] = tsne_features[:, 0]
data['Y'] = tsne_features[:, 1]
```

```python
def recommender(search):
    cs_list = []
    brands = []
    output = []
    binary_list = []

    # Mendapatkan indeks produk yang dicari
    idx = data[data['Product'] == search].index

    # Pastikan hanya ada satu indeks yang ditemukan
    if len(idx) != 1:
        if len(idx) == 0:
            print("Error: Produk tidak ditemukan dalam dataset.")
        else:
            print("Error: Terdapat beberapa produk dengan nama yang sama dalam dataset. Menggunakan indeks pertama yang ditemukan.")
        idx = idx[0] if len(idx) > 0 else None
    else:
        idx = idx[0]

    if idx is None:
        return

    # Mendapatkan binary list untuk produk yang dicari
    point1 = ingred_matrix.iloc[idx, 1:].values.reshape(1, -1)

    # Mendapatkan tipe produk dan merek produk yang dicari
    prod_type = data.at[idx, 'Category']
    brand_search = data.at[idx, 'Brand']

    # Memfilter data berdasarkan tipe produk
    data_by_type = data[data['Category'] == prod_type]

    # Loop untuk menghitung cosine similarity
    for j in range(len(data_by_type)):
        point2 = ingred_matrix.iloc[j, 1:].values.reshape(1, -1)
        dot_product = np.dot(point1, point2.T)
        norm_1 = np.linalg.norm(point1)
        norm_2 = np.linalg.norm(point2)
        cos_sim = dot_product / (norm_1 * norm_2)
        cs_list.append(cos_sim[0, 0])

    # Menambahkan cosine similarity ke dalam DataFrame
    data_by_type['cos_sim'] = cs_list

    # Mengurutkan data berdasarkan cosine similarity
    data_by_type = data_by_type.sort_values('cos_sim', ascending=False)

    # Memfilter produk yang memiliki nama yang sama dengan produk yang dicari
    data_by_type = data_by_type[data_by_type.Product != search]

    # Memilih merek yang direkomendasikan
    for _, row in data_by_type.iterrows():
        brand = row['Brand']
        if len(brands) == 0 or brands.count(brand) < 2:
            if brand != brand_search:
                brands.append(brand)
                output.append(row)

    # Menampilkan hasil rekomendasi
    print('\033[1m', 'Recommending products similar to', search,':', '\033[0m')
    print(pd.DataFrame(output)[['Product', 'cos_sim']].head(5))
```
